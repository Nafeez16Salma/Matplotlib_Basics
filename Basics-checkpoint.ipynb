```python
 #importing package
    
import matplotlib
```

```python
#importing submodule pyplot 
import matplotlib.pyplot as plt

#numpy for mathematical operations
import numpy as np
```

```python
#drawing a line diagram from (0,0) to (10,50)

xpoints= np.array([0,10])
ypoints = np.array([0,50])

plt.plot(xpoints,ypoints)
plt.show()
```

```python
#plotting only a point instead of line diagram

xpoints = np.array([5,15,30,50])
ypoints = np.array([20,10,40,20])

plt.plot(xpoints,ypoints, 'o')
plt.show()
#line plot
plt.plot(xpoints,ypoints)
plt.show()
```

```python
#markers
##marking each point and plotting a line

plt.plot(xpoints,ypoints, '*:')
plt.show()
```

```python
#format strings [color,points,lines,marker size]
plt.plot(xpoints,ypoints, 'bo--', ms= 10)
plt.show()
```

```python
#marker colors [ markersize, edge clr, face clr]
plt.plot(xpoints,ypoints, 'co--', ms=15,mec='b',mfc='hotpink')
plt.show()
```

```python
#multiple lines


X1 = np.array([5,15,30,50])
Y1 = np.array([20,10,40,20])

X2 = np.array([10,5,20,40])
Y2 = np.array([30,25,10,30])
plt.plot(X1,Y1,'c:*')
plt.plot(X2,Y2, marker='o', ms= 15, c='hotpink', mec='k',linewidth=2.5)
plt.show()
```

```python
#Adding labels

plt.plot(xpoints,ypoints, c='hotpink')

plt.title("Dimensions of a Box")
plt.xlabel("Height")
plt.ylabel("width")

plt.show()

```

```python
#Font properties

plt.plot(xpoints,ypoints, c='hotpink')

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'sans','color':'purple','size':15}
font3 = {'family':'fantasy','color':'purple','size':15}


plt.title("Dimensions of a Box", fontdict = font1)
plt.xlabel("Height", fontdict = font2)
plt.ylabel("width", fontdict = font2)


plt.show()
```

```python
#Adding and styling Grids

plt.plot(xpoints,ypoints, c='hotpink')


plt.title("Dimensions of a Box")
plt.xlabel("Height")
plt.ylabel("width")

#plt.grid()
#plt.grid(axis='x')
plt.grid(color = 'purple', linestyle = '--', linewidth = 0.5)
plt.show()

```

```python
#subplot functions


X1 = np.array([5,15,30,50])
Y1 = np.array([20,10,40,20])



plt.subplot(2, 2, 1)
#the figure has 2 row, 2 columns, and this plot is the first plot.
plt.plot(X1,Y1)

X1 = np.array([10,5,20,40])
Y1 = np.array([30,25,10,30])

plt.subplot(2, 2, 2)
#the figure has 2 row, 2 columns, and this plot is the second plot.
plt.plot(X1,Y1,'o:')

plt.subplot(2, 2, 3)
#the figure has 2 row, 2 columns, and this plot is the third plot.
plt.plot(X1,Y1,'o-',c='hotpink')

plt.show()
```

```python
#Adding titles to each subplot

X1 = np.array([5,15,30,50])
Y1 = np.array([20,10,40,20])



plt.subplot(3, 2, 1)
#the figure has 3 row, 2 columns, and this plot is the first plot.
plt.plot(X1,Y1)
plt.title("Dimentions of BOX1")
plt.ylabel("width")

X1 = np.array([10,5,20,40])
Y1 = np.array([30,25,10,30])

plt.subplot(3, 2, 2)
#the figure has 3 row, 2 columns, and this plot is the second plot.
plt.plot(X1,Y1,'o:')
plt.title("Dimentions of BOX2")

plt.subplot(3, 2, 3)
#the figure has 3 row, 2 columns, and this plot is the third plot.
plt.plot(X1,Y1,'o-',c='hotpink')


X1 = np.array([5,15,30,50])
Y1 = np.array([20,10,40,20])

plt.subplot(3, 2, 4)
#the figure has 3 row, 2 columns, and this plot is the fourth plot.
plt.plot(X1,Y1,'o-',c='cyan')


plt.subplot(3, 2, 5)
#the figure has 3 row, 2 columns, and this plot is the fifth plot.
plt.plot(X1,Y1,'o--',c='black')



plt.show()
```

```python
#creating scatter plots

X1 = np.array([5,15,30,50,60,70])
Y1 = np.array([3,10,20,30,40,50])

plt.title("CARS",c="purple")
plt.xlabel("Age of a car")
plt.ylabel("speed of a car")

plt.scatter(X1, Y1, color='cyan')

X2 = np.array([4,12,20,30,50,60])
Y2 = np.array([7,9,10,40,45,50])

plt.scatter(X2, Y2, color='magenta')

#colors= np.array(["red","green","blue","hotpink","black","purple"])
#plt.scatter(X2, Y2, c=colors)

plt.show()
```

```python
#using colourmap

colors = np.array([20,10,40,50,60,70])
ax = plt.axes()
  
# Setting the background color of the plot 
# using set_facecolor() method

ax.set_facecolor("grey")
plt.scatter(X1, Y1, c=colors, cmap='Pastel1')
plt.colorbar()
```

```python
#setting sizes for the bubbles
#alpha for tranparency

X1 = np.array([5,15,30,50,60,70])
Y1 = np.array([3,10,20,30,40,50])
sizes = np.array([50,100,300,500,600,700])

plt.scatter(X1, Y1, c=colors, cmap='Pastel1', s=sizes, alpha = 1.0)
plt.show()

```
